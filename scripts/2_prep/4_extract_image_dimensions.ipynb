```python
## Extract image dimensions and add to CSV

from PIL import Image
import os
import pandas as pd

# Read in CSV
input_path = '/Users/caraappel/Documents/CV4E/oregon_critters/metadata_labels/empty/empty_split_hjagrid.csv' 
#input_path = '/Volumes/Cara_cam_traps/CV4E/data_cleaned/demo2/demo2_empties.csv'
dataframe = pd.read_csv(input_path)
```

```python
# Define function to get image dimensions
def get_image_dimensions(filename):
    with Image.open(filename) as img:
        return img.size
```

```python
# Create lists to store dim
width_output = []
height_output = []

# Extract dimensions
for image in dataframe['Full Path'].values:
    width, height = get_image_dimensions(image)
    width_output.append(width)
    height_output.append(height)
    
# Append to CSV
dataframe['img_width'] = width_output
dataframe['img_height'] = height_output

dataframe.head()
```

```python
# Save new CSV
output_path = input_path.replace('.csv', '_dim.csv')
dataframe.to_csv(output_path)

print('saved file' + output_path)
```
