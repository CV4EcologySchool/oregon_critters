```python
# Predict bounding boxes on validation set using trained YOLO model

import wandb
from ultralytics import YOLO

# Start a W&B run
#wandb.init(project='yolo_0_val_predict')

# Load my pretrained YOLOv8n model
#model = YOLO('/home/cara/oregon_critters/runs/detect/train15/weights/best.pt')
model = YOLO('/home/cara/oregon_critters/runs/detect/1_sc/weights/best.pt')

# Read in image paths for validation images
with open('/home/cara/oregon_critters/sampled_ds_300_val.txt', 'r') as f:
    image_paths = f.read().splitlines()
```

```python
# Iterate through images in val set and run predictions
for image in image_paths:
    full_path = '/home/cara/oregon_critters/' + image
    model.predict(full_path, save=True, save_conf = True, imgsz=640, #save_txt=True 
                  conf=0.1, iou=0.7) 
    #set conf and iou low bc I want to see bad ones too

# Images are saved in folder structure /runs/detect/predict in same location as script
# (so for me: oregon_critters/scripts/runs/detect/predict)

#ran first with iou=0.7 (renamed to /detect/predict_iou07)
#then ran again with iou=0.1

# for new binary model: ran predictions with conf=0.1, iou=0.1

#currently running multiclass model, correct size, conf=0.1, iou=0.7
```

```python
# Now look at some metrics and examples

## outputs from training were stored in /runs/detect/1_sc (for "single class")

# Run 'val' to get additional metrics:

#Validate the model
metrics = model.val(data='/home/cara/oregon_critters/dataset.yaml', 
                    save_json=True, save_hybrid=True, conf=0.001, iou=0.1, split='val') 

metrics.box.map    # map50-95
metrics.box.map50  # map50
metrics.box.map75  # map75
metrics.box.maps   # a list contains map50-95 of each category

#what is this creating and where is it saving it?
```
