```python
import pandas as pd
import os
import shutil
from tqdm import tqdm
```

```python
df = pd.read_csv('metadata_labels/sampled_ds_300.csv')
df.head()
```

```python
df['full_path'].values
```

```python
for file in tqdm(df['full_path'].values):
    
    file_source = '../' + file
    
    file_target = '../upload_300thresh/' + file
    
    os.makedirs(os.path.dirname(file_target), exist_ok=True)

    shutil.copy(file_source, file_target)
    
```
